## ***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

## ***LOADING THE DATASET***

```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

## ***SUMMARIZING THE DATASET***

```python
data = pd.DataFrame(dataset.data,columns=[dataset.feature_names]).join(pd.DataFrame(dataset.target,columns=["(target,)"]))
```

```python
print(data.shape)
data.head()
```

## ***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
x = data.iloc[:,:-1].values
print(x)
y = data.iloc[:,-1].values
print(y)
```

## ***TRAINING THE DATASET***

```python
from sklearn.decomposition import PCA

Model = PCA()

Model.fit_transform(x)
```

## ***FINDING THE VARIANCE RATIO OF ALL THE FEATURES***

```python
Model.explained_variance_ratio_*100
```

## ***TRAINING THE DATASET AND PREDICTING THE RESULT USING THE OPTIMAL NUMBER OF COMPONENTS (DIMENSIONS) BASED ON THE VARIANCE RATIO RESULT***

```python
from sklearn.decomposition import PCA

PCA_Model = PCA(n_components=2)

y_pred    = PCA_Model.fit_transform(x)
```

```python
y_pred
```

## ***VISUALIZING THE CLUSTERS***

```python
import matplotlib.pyplot as plt

plt.scatter(y_pred[y==0,0],y_pred[y==0,1],s=100,label='setosa')
plt.scatter(y_pred[y==1,0],y_pred[y==1,1],s=100,label='versicolor')
plt.scatter(y_pred[y==2,0],y_pred[y==2,1],s=100,label='virginica')
plt.title("IRIS CLUSTERING")
plt.legend()
```
